```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\mulka\OneDrive\Documents\Python Scripts\MyData.csv')
df.head(10)
```

```python
# the first 7 rows are dummy rows and we can drop them

df = df.drop(df.index[0:7])
df.head(5)
```

```python
# the first row (index 7) should actually be the column titles

df.columns = df.iloc[0]
df.head(3)
```

```python
# creating a copy of the dataframe
df1 = df.copy()
```

```python
# creating a dictionary to check index positions of each column

index_dic = {}
for col in df1.columns:
    index_dic[col] = df.columns.get_loc(col)
print(index_dic)
```

```python
# we can drop: 'nan', '(1)Comments', '(2)TotalCharge', '(2)Comments','$Difference','%Difference'

df1 = df1.drop(df1.columns[[10, 13, 14, 15, 16]], axis=1)
df1.head()
```

```python
df1.columns.tolist()
```

```python
df1 = df1.rename(columns={'\ufeffOrigin City': 'Origin City'})
df1.head()
```

```python
# removing the duplicate title column

df1 = df1.drop(df1.index[0])
df1.head(3)
```

```python
# checking that the re-name worked

df1.columns.tolist()
```

```python
# checking data types

df1.dtypes
```

```python
# Total Charge and Avg Weight need to be changed to numeric values

df1['(1)TotalCharge'] = pd.to_numeric(df1['(1)TotalCharge'])
df1['Avg Weight (LBS)'] = df1['Avg Weight (LBS)'].astype(float)
df1.dtypes
```

```python
# Creating a new column that gives us the dry van shipping rate (the current rate is refrigerated truck, 
# so we divide that rate by 2 to get the rate for a dry van)

df1['Total Dry Van Rate'] = df1['(1)TotalCharge']/2
df1['Dry Van Rate per LB'] = df1['Total Dry Van Rate']/df1['Avg Weight (LBS)']
df1.head()
```

```python
# Getting the Refrigerated Truck cost per LB

df1['Reefer Rate per LB'] = df1['(1)TotalCharge']/df1['Avg Weight (LBS)']
df1.head()
```

```python
# doing a couple last checks to make sure the data is accurate

df1.info()
```

```python
# checking for nulls

df1.isnull().any()
```

```python
df1 = df1.rename(columns={'(1)TotalCharge': 'Total Reefer Rate', '(1)Comments': 'Output Truck Type'})
df1.columns.tolist()
```

```python
df1.to_csv('Software Output.csv', index=True)
```
